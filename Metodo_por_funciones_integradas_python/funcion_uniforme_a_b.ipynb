# Variables aleatorios uniformes en el intervalo [a,b]

```python
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
a = -2.0
b = 6
unif_a_b =[]
iteraciones = 10000
for i in range(0,iteraciones):
    if a < b:
        unif_a_b.append( a + ((b-a)*random.random()))
    else:
        print(" ################# Intervalo no valido #################")
        break
array_uniforme = np.array(unif_a_b)
mean_unif =array_uniforme.mean()
var_unif = (iteraciones / (iteraciones-1))*np.var(array_uniforme)
print("El promedio de las variables aleatorias es: {0:0.8f}".format(mean_unif))
print("La varianza de las variables aleatorias es: {0:0.8f}".format(var_unif))

plt.figure(figsize=(10,6))
plt.hist(array_uniforme, color = 'darkcyan', edgecolor = "black", linewidth=2 , density = True)
plt.title("Funcion uniforme en el intervalo [a,b]",fontsize=23)
plt.xlabel("Clases", fontsize=15)
plt.ylabel("Frecuencia", fontsize=15)
```
