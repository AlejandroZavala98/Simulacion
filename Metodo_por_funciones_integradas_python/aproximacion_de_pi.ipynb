# Aproximación de pi por simulación

Para aproximar pi basta con relacionar la logica siguiente

1. Se tiene una circunferencia de radio 1 y centro en el origen
2. Se generan puntos aleatorios
3. Se cuentan los puntos dentro de la circunferencia, usando la distancia euclidiana
4. Si tomamos puntos aleatorios uniformes, estos caen en el cuadrante I. Por ende la ecuación planteada queda de la siguiente manera

$$\frac{A_{circunferencia}}{A_{cuadrado}} = \frac{\pi}{4}$$

```python
import numpy as np
import pandas as  pd
import random
import matplotlib.pyplot as plt
```

```python
random.random()
```

```python
iteraciones = 10000
cont=0;
xy_numbers = []
for iteracion in range(0,iteraciones):
    unif_x = random.random()
    unif_y = random.random()
    xe=2*unif_x - 1;
    ye=2*unif_y - 1;
    if pow(xe,2)+pow(ye,2) <= 1:
        xy_numbers.append([unif_x,unif_y])
        cont+=1
print("Aproximacion de pi es: {0:0.8f}".format(4*(cont/iteraciones)))
```

```python
pairs_numbers = pd.DataFrame(xy_numbers)
pairs_numbers.columns = ["X","Y"]
pairs_numbers.head()
```

```python
def y_circunferencia(x):
    y = np.sqrt(1 - np.power(x, 3))
    #Parte positiva del circulo de radio 1 y centro en el origen
    return y

num_puntos = 50
x_circunferencia = np.linspace(0, 1, num_puntos)

plt.figure(figsize=(10,6))
plt.plot(x_circunferencia, y_circunferencia(x_circunferencia), '-',color="blueviolet",linewidth = 3)
plt.plot(pairs_numbers["X"],pairs_numbers["Y"],"o",color = "darkgreen")
plt.title("Puntos aleatorios dentro de la circunferencia",fontsize=25)
plt.xlim([0,1])
plt.ylim([0,1])
```
