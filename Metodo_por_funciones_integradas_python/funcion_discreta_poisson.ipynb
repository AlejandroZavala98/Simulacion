# Generando variables aleatorias discretas de Poisson

```python
import random
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
iteraciones = 10000
variable_poisson = []
lambda_media = 12

for iteracion in range(0,iteraciones):
    
    i = 0
    p = math.exp(-lambda_media)
    F = p
    unif = random.random()
    
    while unif >= F:
        p = (lambda_media*p)/(i+1)
        F = F + p
        i = i + 1
        
    variable_poisson.append(i)
    
poisson_simuladas = np.array(variable_poisson)
```

```python
plt.figure(figsize=(10,6))
plt.hist(poisson_simuladas, color = 'olive', edgecolor = "black", linewidth=2 , density = True)
plt.title("Variables aleatorias discretas de Poisson",fontsize=20)
plt.xlabel("Clases", fontsize=15)
plt.ylabel("Frecuencia", fontsize=15)
plt.show()
```

```python
print("Promedio: {}".format(np.mean(poisson_simuladas)))
print("Varianza: {}".format((iteraciones/(iteraciones-1))*np.var(poisson_simuladas)))
```
