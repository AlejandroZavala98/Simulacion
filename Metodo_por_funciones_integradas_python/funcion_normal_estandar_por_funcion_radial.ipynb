# Generando variables normales de media cero y varianza 1 con funcion radial

```python
import math
import numpy as np
import pandas as  pd
import random
import matplotlib.pyplot as plt
```

```python
x_normal = []
y_normal = []

iteraciones=10000

for iteracion in range(0,iteraciones):
    variable_0_2pi = 2*math.pi*random.random()
    variable_radial = np.sqrt(-2*np.log(random.random()))
    x_normal.append(variable_radial*np.cos(variable_0_2pi))
    y_normal.append(variable_radial*np.sin(variable_0_2pi))
x_normales = np.array(x_normal)
y_normales = np.array(y_normal)
```

```python
plt.figure(figsize=(10,6))
plt.hist(x_normales, color = 'darkcyan', edgecolor = "black", linewidth=2 , density = True)
plt.title("Variables aleatorias normales",fontsize=20)
plt.xlabel("Clases", fontsize=15)
plt.ylabel("Frecuencia", fontsize=15)
plt.show()
```

```python
print("Promedio: {}".format(np.mean(x_normales)))
print("Varianza: {}".format((iteraciones/(iteraciones-1))*np.var(x_normales)))
```

```python
plt.figure(figsize=(10,6))
plt.hist(y_normales, color = 'darkred', edgecolor = "black", linewidth=2 , density = True)
plt.title("Variables aleatorias normales",fontsize=20)
plt.xlabel("Clases", fontsize=15)
plt.ylabel("Frecuencia", fontsize=15)
plt.show()
```

```python
print("Promedio: {}".format(np.mean(y_normales)))
print("Varianza: {}".format((iteraciones/(iteraciones-1))*np.var(y_normales)))
```
